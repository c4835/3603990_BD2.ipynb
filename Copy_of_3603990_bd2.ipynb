### 2.3 Price Cleaning in Product.csv and product.json
###### As there are more than 27% of values missing in both product json and csv datasets in the column of "price" and "list_price" its very important to clean them and try to fill them with sensible logic which will help in further analysis for more accurate visuals. And as we have previously noticed that there are same number of values missing in both 2 columns of both datasets, lets try to match them now to validate our observation.

###### 2.3.1 Here we will compare price column from product.csv and list_price column from products.json

```python
# Here we will merge the price from csv and list_price from json in one temporary dataset toc check how closely related they both are.
# We will loose loc function for slicing and merging the colums from 2 seperate databased.

price_compare = products.loc[:,['uniq_id','price']].merge(products_json.loc[:,['uniq_id','list_price']], on='uniq_id', how='inner')
price_compare.head(2)
```

###### 2.3.2 Now as we have established price and list_price are identicle, lets check the correlation between list_price and sale_price column in products.json file.

```python
# Here we will merge the list_price and sale_price from json in one temporary dataset
products_json_float = products_json.loc[:,['uniq_id','sale_price', 'list_price']]
products_json_float.head(2)
```

```python
# lets check if any negative values present in list_price or sale_price

print((products_json_float['sale_price']<0).sum())
print((products_json_float['list_price']<0).sum())
```

```python
products_json_float[['sale_price','list_price']].corr()
```

###### As we can see here the correlation coefficient in between two columns is not as close. we will first find out how much is the average discount given comparing sale price and list price.

```python
#As we can see here the correlation coefficient in between two columns is not very close.
#we will first find out how much is the avergae discount given comparing sale price and list price.

products_json_float['discount'] = products_json_float['list_price'] - products_json_float['sale_price']
products_json_float['discount_percentage'] = ((products_json_float['discount'] / products_json_float['list_price']))*100
products_json_float['discount_percentage'].head(2)
```

```python
# Also lets calclulate the average discount given from the entire data
average_discount = products_json_float['discount_percentage'].mean()    # '.mean' will give a mean (average value)
average_discount = round(average_discount, 2)                           # 'round' will limit the final number to 2 decimal places
print(f"The average discount is: {average_discount}%")
```

```python
# Now lets check compare the each missing values rows of list_price which have full value in sale_price and make seperate dataset for it
null_rows = products_json_float.loc[products_json_float['list_price'].isnull()]
print (null_rows.shape)
print (null_rows.head(2))
print (null_rows.isnull().sum())
```

```python
# Now will drop the null values of sale_price(as they are very low in proportion compared to list_price) and store the dataset in new dataset
sale_drop = null_rows.dropna(subset=['sale_price'])    # 'dropna' is used to remove the rows which have missing values
print (sale_drop.isnull().sum())
print (sale_drop.shape)
```

```python
# Now using filna, the 'Average discount given: 41.58%' to the every sale row in 'sale_drop' and impute its adjecent row with this added value.
# Lets do that using function

# We will first create a copy of product_json and impute the values in them
products_json_imp = products_json.copy()

def cal_list(sale_price, discount_rate):    # This defines the function which calculate the number based on sale row value and the avrage dicsount
  if pd.notna(sale_price):
    return sale_price / (1 - discount_rate)
  else:
    return np.nan
discount_rate = 0.41

# Loop to check ebery row to fill missing list_prices
# It also checks if 'list_price' is missing and adjescent 'sale_price' exists
# then it calculates the 'list_price' by using the formula created in previous function
for i in products_json_imp.index:
  if pd.isna(products_json_imp.at[i, 'list_price']) and pd.notna(products_json_imp.loc[i, 'sale_price']):
    products_json_imp.loc[i, 'list_price'] = cal_list(products_json_imp.loc[i, 'sale_price'],discount_rate)

print(products_json_imp[['list_price','sale_price']].isnull().sum())
print (products_json_imp.shape)
products_json_imp.tail(2)
```

```python
# Now lets drop the null values in product_json_imp
# As the remaining null values in list_price and sale_price are comparitively less after imputation
products_json_imp = products_json_imp.dropna(subset=['list_price', 'sale_price'])    #'subset' will only use specified colums for operation
print (products_json_imp.shape)
products_json_imp[['list_price','sale_price']].isnull().sum()
```

###### To check if our imputation formula gave a closely accurate result, we will check how close the results are comparing them to the prices of sale and list prices where both were existed before the imputation. We can do that by calculating Mean Absolute Error (MAE)

```python
# To do that we will import the library which will be to caluclate mean absolute error
from sklearn.metrics import mean_absolute_error

# Will select the prices where both list_price and sale_price exists before imputation
check_df = products_json.loc[products_json['list_price'].notnull() & products_json['sale_price'].notnull(), ['list_price', 'sale_price']]

# We will take sample (10%) to validate (frac=0.1)
check_df = check_df.sample(frac=0.1, random_state=40)    # random_state is for choosing random sample

# We will store the actual values
actual_list_price = check_df['list_price']
actual_sale_price = check_df['sale_price']

# Will apply our imputation logic
cal_list_price = check_df['sale_price'] / (1 - discount_rate)

# Calculate Error
mae_list_price = mean_absolute_error(actual_list_price, cal_list_price)
mae_sale_price = mean_absolute_error(actual_sale_price, check_df['list_price'])

print(f"MAE for list_price: {mae_list_price}")
```

###### Considering wide range of price variations in the datasets, MAE of $32 variation in accuracy is moderate.

###**Note:**
######I did not use the usual statistical mean or median method of imputation as sale_price and list_price colums are highly corelated with the average percentage of discount between each other, thus I preferred this method for more realistic prices where the available sale price of every product is used to calculate the list price by using the average discount.

### **5.3 Average ratings by age groups**

```python
# Lets first define age groups
age_bins = [0, 18, 35, 50, 65, 100]    # Define numeric ranges that will be used to split ages into groups
age_labels = ['<18', '18-34', '35-49', '50-64','65+']    # Create labels that will represent each age range

# Create a new column 'Age_Group' by categorising the 'age' column using the defined bins and labels
merged_df['Age_Group'] = pd.cut(merged_df['age'], bins=age_bins, labels=age_labels, right=False)
# "right=False" means bins are left-inclusive and right-exclusive

# Calculate average rating per age group
age_group_ratings = merged_df.groupby("Age_Group")["score"].mean().sort_index()    # "sort_index()" It means sort the groups in order.

plt.figure(figsize=(7,4))

plt.plot(age_group_ratings.index.astype(str), age_group_ratings.values)    # "astype(str)" converts the index strings into lables

plt.xlabel("Age Group")
plt.ylabel("Average Rating")
plt.title("Average Rating by Age Group")

plt.show()
```
