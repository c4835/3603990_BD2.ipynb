```python
# Here we will merge the list_price and sale_price from json in one temporary dataset
products_json_float = products_json.loc[:,['uniq_id','sale_price', 'list_price']]
products_json_float.head(2)
```

###### As we can see here the correlation coefficient in between two columns is not as close. we will first find out how much is the average discount given comparing sale price and list price.

```python
#As we can see here the correlation coefficient in between two columns is not very close.
#we will first find out how much is the avergae discount given comparing sale price and list price.

products_json_float['discount'] = products_json_float['list_price'] - products_json_float['sale_price']
products_json_float['discount_percentage'] = ((products_json_float['discount'] / products_json_float['list_price']))*100
products_json_float['discount_percentage'].head(2)
```

```python
# Also lets calclulate the average discount given from the entire data
average_discount = products_json_float['discount_percentage'].mean()    # '.mean' will give a mean (average value)
average_discount = round(average_discount, 2)                           # 'round' will limit the final number to 2 decimal places
print(f"The average discount is: {average_discount}%")
```

```python
# Now lets check compare the each missing values rows of list_price which have full value in sale_price and make seperate dataset for it
null_rows = products_json_float.loc[products_json_float['list_price'].isnull()]
print (null_rows.shape)
print (null_rows.head(2))
print (null_rows.isnull().sum())
```

```python
# Now will drop the null values of sale_price(as they are very low in proportion compared to list_price) and store the dataset in new dataset
sale_drop = null_rows.dropna(subset=['sale_price'])    # 'dropna' is used to remove the rows which have missing values
print (sale_drop.isnull().sum())
print (sale_drop.shape)
```

```python
# Now using filna, the 'Average discount given: 41.58%' to the every sale row in 'sale_drop' and impute its adjecent row with this added value.
# Lets do that using function

# We will first create a copy of product_json and impute the values in them
products_json_imp = products_json.copy()

def cal_list(sale_price, discount_rate):    # This defines the function which calculate the number based on sale row value and the avrage dicsount
  if pd.notna(sale_price):
    return sale_price / (1 - discount_rate)
  else:
    return np.nan
discount_rate = 0.41

# Loop to check ebery row to fill missing list_prices
# It also checks if 'list_price' is missing and adjescent 'sale_price' exists
# then it calculates the 'list_price' by using the formula created in previous function
for i in products_json_imp.index:
  if pd.isna(products_json_imp.at[i, 'list_price']) and pd.notna(products_json_imp.loc[i, 'sale_price']):
    products_json_imp.loc[i, 'list_price'] = cal_list(products_json_imp.loc[i, 'sale_price'],discount_rate)

print(products_json_imp[['list_price','sale_price']].isnull().sum())
print (products_json_imp.shape)
products_json_imp.tail(2)
```

```python
# Now lets drop the null values in product_json_imp
# As the remaining null values in list_price and sale_price are comparitively less after imputation
products_json_imp = products_json_imp.dropna(subset=['list_price', 'sale_price'])    #'subset' will only use specified colums for operation
print (products_json_imp.shape)
products_json_imp[['list_price','sale_price']].isnull().sum()
```

###**Note:**
######I did not use the usual statistical mean or median method of imputation as sale_price and list_price colums are highly corelated with the average percentage of discount between each other, thus I preferred this method for more realistic prices where the available sale price of every product is used to calculate the list price by using the average discount.

###### In a scatterplot below, a strong positive correlation is observed in list and sale price, we can see that sale prices increase proportionally with list prices. colour coding is applied to discount groups, it also shows that most products fall within the 10 to 30% discount category, while a smaller number receive larger discounts exceeding 50%. It shows consistent pricing & discount strategy across the product range.

```python
# Lets make discount percentage groups of 0-10%, 10-30%, 30-50%, 50%+.

# First we will create a new column by subtracting sale price from list price
merged_df['discount'] = merged_df['list_price'] - merged_df['sale_price']

# This will calculate the discount percentage of each product
merged_df['discount_percent'] = (
    merged_df['discount'] / merged_df['list_price']
) * 100


merged_df['discount_group'] = pd.cut(            # "pd.cut" will split the continues numeric data into defined ranges
    merged_df['discount_percent'],
    bins=[0,10,30,50,100],                       # "Bins" will define the numerical ranges to split the data
    labels=['0-10%','10-30%','30-50%','50%+']    # "lables" will assign names to each range.
)

plt.figure(figsize=(8,4))

# "hue" will assign colours to the discount ranges
# "palette" will choose the color scheme "viridis" for particular set of colours
sns.scatterplot(x='list_price', y='sale_price', data=merged_df, hue='discount_group', palette='viridis')
plt.title('List Price vs. Sale Price with discount groups')
plt.xlabel('List_Price')
plt.ylabel('Sale_Price')

plt.show()
```
